```python
import sys, os, importlib, pkgutil

def where(modname: str):
    try:
        m = importlib.import_module(modname)
    except Exception as e:
        print(f"{modname}: IMPORT ERROR -> {e}")
        return
    origin = getattr(m, "__file__", None)
    spec   = getattr(m, "__spec__", None)
    pth    = getattr(m, "__path__", None)
    print(f"{modname}: file={origin!r}, origin={getattr(spec,'origin',None)!r}")
    if pth is not None:
        print(f"         package path entries: {list(pth)}")

for name in ("roboguard", "spine", "spot"):
    where(name)

print("python:", sys.executable)
print("CWD    :", os.getcwd())
print("sys.path[0:3]:", sys.path[:3])
```

```python
import sys, importlib, pkgutil
for name in ("spine","roboguard","spot"):
    try:
        m = importlib.import_module(name)
        print(f"{name}: origin={getattr(m.__spec__,'origin',None)!r}, file={getattr(m,'__file__',None)!r}")
        if getattr(m, "__path__", None):
            print("   package path entries:", list(m.__path__))
    except Exception as e:
        print(f"{name}: IMPORT ERROR -> {e}")
```
